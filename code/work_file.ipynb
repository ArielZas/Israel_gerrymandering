```python
right_block = ['מחל', 'שס', 'ג', 'ט', 'ב']
left_block = ['אמת', 'מרצ', 'כן', 'פה', 'ל']
arab_block = ['ד', 'ום', 'עם']

results_df["left_votes"] = results_df[left_block].sum(axis=1)
results_df["right_votes"] = results_df[right_block].sum(axis=1)
results_df["arab_votes"] = results_df[arab_block].sum(axis=1)

# group results together 
results_grouped_df = (
    results_df.groupby(["yisuv id", "rikuz"])
    .agg(
        left=("left_votes", "sum"),
        right=("right_votes", "sum"),
        arab=("arab_votes", "sum")
    )
    .reset_index()
)

#  group locations together 
locations_grouped_df = (
    locations_df.groupby(["yisuv id", "rikuz"])
    .agg(
        population=("eligible", "sum"),
    )
    .reset_index()
)


print(results_grouped_df.head())
print(locations_grouped_df.head())
```
